```python
import numpy as np
from copy import deepcopy

x = [1, 2, 5, -4, 6, 8]
y = [2, 2, 2, 2, 2, 2]
z = [-1, -2, -5, 4, -6, -8]
```

```python
def sphere(x):
  total = 0
  for i in x:
    total += i*i
  return total
```

```python
class problem:
  def __init__(self):
    self.number_of_genes = 8
    self.min_value = -10
    self.max_value = 10
    self.cost_function = sphere
    pass
```

```python
class individual:
  def __init__(self, problem):
    self.chromosome = np.random.uniform(problem.min_value, problem.max_value, problem.number_of_genes)
    self.cost = problem.cost_function(self.chromosome)

  def mutate(self, rate_of_gene_mutation, range_of_gene_mutation):
    for index in range(len(self.chromosome)):
      if np.random.uniform() < rate_of_gene_mutation:
        self.chromosome[index] += np.random.rand() * range_of_gene_mutation

  def crossover(self, parent2, explore_crossover):
    alpha = np.random.uniform(-explore_crossover, 1 + explore_crossover)
    child1 = deepcopy(self)
    child2 = deepcopy(parent2)
    child1.chromosome = alpha*self.chromosome + (1-alpha)*parent2.chromosome
    child2.chromosome = alpha*parent2.chromosome + (1-alpha)*self.chromosome
    return child1, child2
```

```python
def choose_indices_from(number_in_list):
  index_1 = np.random.randint(number_in_list)
  index_2 = np.random.randint(number_in_list)
  if index_1 == index_2:
    return choose_indices_from(number_in_list)
  return index_1,index_2
```

```python
class parameters:
  def __init__(self):
    self.population_size  = 1000

    self.max_generation = 1000

    self.crossover_explore_rate = 0.2
    self.gene_mutate_rate = 0.2
    self.gene_mutate_range = 1

    self.birthrate_per_generation = 1
```

```python
def choose_parents(population_size):
  index1 = np.random.randint(0, population_size)
  index2 = np.random.randint(0, population_size)
  if index1 == index2:
    return choose_parents(population_size)
  else:
    return index1, index2
```

```python
def run_genetic(problem, params):
  """
  1. Read Variables
  2. Create Population
  3. Training Loop
  start loop
    4. Create Children
    5. Sort Population
    6. Cull Population - choose parent, use it to find more children
    7. Check Solution
  end loop

  How to effectively choose parents (roulette, tournament)
  """
  # Cost Function
  cost_function = problem.cost_function
  acceptable_cost = problem.acceptable_cost

  # Params

  population_size  = params.population_size

  max_generation = params.max_generation

  explore_crossover = params.crossover_explore_rate
  gene_mutate_rate = params.gene_mutate_rate
  gene_mutate_range = params.gene_mutate_range

  birthrate_per_generation = params.birthrate_per_generation

  population = []

  #placeholder for best solution
  best_solution = individual(problem)
  best_solution.cost = np.infty

  # 1. Create our population
  for i in range(population_size):
    new_individual = individual(problem)
    population.append(new_individual)

    if new_individual.cost < best_solution.cost:
      best_solution = deepcopy(new_individual)

  # 3. Create a loop
  for i in range(max_generation):
    children = []
    while len(children) < birthrate_per_generation:
    # 4. Create Children
    # choose parents
      parent1_index, parent2_index = choose_parents(population_size)
      parent1 = population[parent1_index]
      parent2 = population[parent2_index]

      # crossover
      child1, child2 = parent1.crossover(parent2, explore_crossover)

      # mutate
      child1.mutate(gene_mutate_rate, gene_mutate_range)
      child2.mutate(gene_mutate_rate, gene_mutate_range)

      # evaluate cost
      child1.cost = cost_function(child1.chromosome)
      child2.cost = cost_function(child2.chromosome)

      # add children to children
      children.append(child1)
      children.append(child2)

      population.extend(children)

      # 5. Sort Population
      population.sort(key=lambda x: x.cost)

      # 6. Cull Population
      population = population[:population_size]
      print(population[0].cost)

      # 7. Check solution
      if population[0].cost < best_solution.cost:
        best_solution = deepcopy(population[0])
        print(best_solution.cost)

      if best_solution.cost < acceptable_cost:
        break

  return population, best_solution
```

```python
def sumPows2(v):
  targetValue = 1
  sum = 0
  for x in v:
    sum += (x - targetValue)**2
    targetValue *= 2
  return sum
```
